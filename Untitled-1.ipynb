```python
# Grouping Method 1  4 Groups of 3 people are required 1F/2L ,7 Groups of 2 people are requured 1F/1L.
# Use "if" Check for compliance with 1F/2L
def Grouping1(s: list,f: list):
#Declare variables
    stack = []                      #Name
    stack1 = []                     #Scores
    stack2 = []                     #F/L
    Rstack =[]                      #Return Name
    Rstack1 = []                    #Return Scores
    Rstack2 = []                    #Return F/L
    Avgstack = []                   #Return Avg
    Min = 0
    Sum = 0.0
    Avg = 0
    CacheData = 0
    i = -1
#Take FData every loop The Highest scores element than store to stack, stack1, stack2. Until group allocation is completed
    for c in s:
        i =  i + 1
        stack.append(c[3])
        stack1.append(c[2])
        stack2.append(c[1])
#Take LData every loop The Lowest scores element than store to stack, stack1, stack2
        for p in f:
            Min = p[1]
        stack.append(p[3])
        stack1.append(p[2])
        stack2.append(p[1])
#If The first 4 Groups have been separated ,start separat 7 Groups
        if i >= 4:
            for p in f:
                Sum = stack2[0] + Min
                Avg = Sum/2
 
        CacheData = [item for item in f if item[1] == Min and item [2] == p[2]]
        f.remove(CacheData[0])
#If The first 4 Groups not separated ,start separat first 4 Groups
        if i < 4:
            for p in f:
                Sum = stack2[0] + stack2[1] + p[1]
                Avg = Sum/3
#If Avg is in the range 86 to 74
                if 6 > (Avg - 80) > -6  :
                    stack.append(p[3])
                    stack1.append(p[2])
                    stack2.append(p[1])
#Avoid deleting data due to repeated selection
                    f.remove(p)
                    break
                else:
                    continue
 
#Store information in the Rstack, Rstack1, Rstack2, Avgstack
        Rstack.append(stack)
        Rstack1.append(stack1)
        Rstack2.append(stack2)
#Initialize variables-------------
        Avg = ("%.2f"%Avg)
        Avgstack.append(Avg)
        stack = []
        stack1 = []
        stack2 = []
        Min = 0
        Sum = 0.0
        Avg = 0
        CacheData = 0
#Initialize variables------------
#Print Group,F/L, everyone of Group scores ,Group Avg.
        print(Rstack1[i],end = ' ')
        print(Rstack[i], ":", end=' ')
        print("Scores",Rstack2[i],end = ' ')
        print("Group Avg."+Avgstack[i])
 
# Data It's from the professor's Excel file.
# (1, 99.0, 'A', 'F')-(26, 44.4, 'Z', 'L')
Data = [
    (1, 99.0, 'A', 'F'),
    (2, 98.0, 'B', 'L'),
    (3, 97.8, 'C', 'F'),
    (4, 97.4, 'D', 'F'),
    (5, 97.3, 'E', 'F'),
    (6, 96.4, 'F', 'L'),
    (7, 96.3, 'G', 'L'),
    (8, 95.4, 'H', 'F'),
    (9, 94.9, 'I', 'F'),
    (10, 93.8, 'J', 'F'),
    (11, 93.5, 'K', 'L'),
    (12, 93.5, 'L', 'L'),
    (13, 93.4, 'M', 'L'),
    (14, 93.3, 'N', 'F'),
    (15, 93.1, 'O', 'L'),
    (16, 92.4, 'P', 'L'),
    (17, 91.8, 'Q', 'L'),
    (18, 89.8, 'R', 'L'),
    (19, 82.5, 'S', 'F'),
    (20, 81.6, 'T', 'L'),
    (21, 70.6, 'U', 'L'),
    (22, 69.6, 'V', 'L'),
    (23, 68.9, 'W', 'F'),
    (24, 58.0, 'X', 'F'),
    (25, 57.1, 'Y', 'L'),
    (26, 44.4, 'Z', 'L')
]
# Separate local students from international students and then arrange them based on their academic performance.
FData = [item for item in Data if item[3] == 'F'  ]
LData = [item for item in Data if item[3] == 'L'  ]
Grouping1(FData,LData)
```
